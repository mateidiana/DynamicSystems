```python
#Aufgabe 3
t,r,x0,r0=var('t,r,x0,r0')
x=function('x')(t)
deqM=diff(x,t)==r*x
sM(t,x0,r)=desolve(deqM,[x,t],[0,x0])
sM
```

```python
solM(t,x0,r)=sM
solM
```
